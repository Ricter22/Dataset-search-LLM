````python
def call_openai_api(prompt, model):
    client = OpenAI()

    if model == 'o1-mini':
        messages = [
            {
                "role": "user",
                "content": prompt
            }
        ]
    else:
        messages = [
            {"role": "system", "content": "You are a helpful assistant."},
            {
                "role": "user",
                "content": prompt
            }
        ]

    completion = client.chat.completions.create(
    model=model,
    messages=messages
)
    return completion
````

````python
def generate_prompt(subqueries):
   PROMPT = f"""
   ## Objective
   Evaluate the quality of query decomposition, assessing how effectively the original complex query is broken down into meaningful, manageable sub-queries.
   In our system we process datasets and from them generate possible user queries.
   The subqueries you will evaluate are to be used in a retrieval system to find relevant datasets and will be matched against the dataset queries generated by us.   

   ## Evaluation Criteria
   Rate each criterion on a scale from 1 (lowest) to 10 (highest):

   1. **Decomposition Clarity**
      - How clearly and logically are the sub-queries separated?
      - Assess the readability and immediate understanding of each decomposed query.
      - Consider whether each sub-query is self-contained and meaningful.

   2. **Comprehensive Coverage**
      - Do the sub-queries collectively cover all aspects of the original query?
      - Evaluate the completeness of the decomposition in addressing the full scope of the initial request.
      - Check if any critical elements of the original query are missed.

   3. **Granularity**
      - Are the sub-queries appropriately sized?
      - Evaluate whether the decomposition is too granular (overly complex) or too broad (lacking detail).
      - Consider the balance between simplicity and depth of each sub-query.

   4. **Executability**
      - How feasible are the sub-queries in terms of being answered or investigated?
      - Assess the practical ability to find information or derive insights for each sub-query.
      - Consider the specificity and actionability of each decomposed query.

   5. **Context Preservation**
      - Do the sub-queries maintain the original intent and context of the main query?
      - Evaluate how well the decomposition preserves the core meaning and nuance.
      - Check that no critical contextual information is lost in the breakdown.

   ## Evaluation Process
   For each criterion:
   - Provide a score from 1-10
   - Include a brief justification explaining the rationale behind the score
   - Offer specific suggestions for improvement where applicable

   ## Output Format
   {OUTPUT}

   ## Evaluation Inputs
   - **Original Query**: {QUERY}
   - **Decomposed Queries**: {subqueries}

   ## Additional Guidance
   - Consider the domain and complexity of the original query
   - Assess the decomposition from the perspective of an expert researcher or information seeker
   - Evaluate the potential for efficient information gathering based on the sub-queries

   ## Final Recommendation
   Provide a comprehensive assessment of the query decomposition's effectiveness, highlighting strengths and potential areas for improvement.
   """
   return PROMPT
````

````python
import json

def print_json_content(json_response):
    json_response = json_response.replace("```json\n", "").replace("\n```", "")

    json_response_dict = json.loads(json_response)
    
    # Stampa ogni elemento del JSON in modo ordinato
    for key, value in json_response_dict.items():
        print(f"{key.replace('_', ' ').title()}:")
        
        # Se il valore è un dizionario, stampa i suoi sotto-elementi
        if isinstance(value, dict):
            for sub_key, sub_value in value.items():
                print(f"  {sub_key.replace('_', ' ').title()}: {sub_value}")
        else:
            print(f"  {value}")
        print()  # Riga vuota per separare gli elementi
````

````python
evaluation = call_openai_api(generate_prompt(SUMMARY_AUG), 'o1-mini')
````

````python
print_json_content(evaluation.choices[0].message.content)
````

````python
evaluation_kp = call_openai_api(generate_prompt(KP_AUG), 'o1-mini')

print_json_content(evaluation_kp.choices[0].message.content)
````

````python
evaluation_kp_summary = call_openai_api(generate_prompt(SUMMARY_KP_AUG), 'o1-mini')

print_json_content(evaluation_kp_summary.choices[0].message.content)
````
