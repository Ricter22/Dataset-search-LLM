A questo punto prepariamo il sample, nel nostro caso sarà un JSON.




````python
def sample_csv_for_llm(csv_path: str, sample_size: int = 500):

    df = pd.read_csv(csv_path)

    if len(df) < sample_size:
        sample_size = len(df)

    sample_df = df.sample(n=sample_size, random_state=42)
    sample_json = sample_df.to_json(orient='records', date_format='iso')

    return sample_json
````

Now we setup the judge and its prompt

````python
from dotenv import load_dotenv
from openai import OpenAI
````

````python
#setup OpenAI API
load_dotenv()
client = OpenAI()
````

````python
def call_openai_api(prompt, model):

    if model == 'o1-mini':
        messages = [
            {
                "role": "user",
                "content": prompt
            }
        ]
    else:
        messages = [
            {"role": "system", "content": "You are a judge which is expert in evaluating LLM generation."},
            {
                "role": "user",
                "content": prompt
            }
        ]

    completion = client.chat.completions.create(
    model=model,
    messages=messages
)
    return completion
````

````python
def generate_prompt(prompt, output):

  PROMPT = f"""
  ## Objective
  You will be given the input prompt and the output generated by a language model (LM) tasked with providing user instructions for discovering a specific dataset. Your goal is to evaluate the quality of user instructions generated for discovering a specific dataset, focusing on their effectiveness, comprehensiveness, and usability.

  ## Input
  Prompt: {prompt}
  Output: {output}

  ## Evaluation Criteria
  Rate each criterion on a scale from 1 (lowest) to 10 (highest):

  1. **Recall**
    - How comprehensively do the instructions enable users to understand and access the key components of the dataset?
    - Assess the clarity and completeness of guidance for navigating the dataset's structure and content.

  2. **Precision**
    - How accurately do the instructions direct users to the most relevant and important aspects of the dataset?
    - Evaluate the instructions' ability to filter out irrelevant paths and focus on meaningful data exploration.

  3. **Specificity**
    - How precisely do the instructions outline specific methods, tools, or approaches for dataset analysis?
    - Consider the level of detail in suggesting concrete exploration strategies.

  4. **Breadth**
    - To what extent do the instructions cover different potential use cases and analysis approaches?
    - Assess the range of exploratory paths and analytical perspectives suggested.

  5. **Depth**
    - How thoroughly do the instructions provide insights into advanced or nuanced exploration techniques?
    - Evaluate the depth of analytical suggestions and the potential for discovering complex patterns.

  6. **Relevance**
    - How well do the instructions align with potential user goals and research interests?
    - Consider the adaptability and applicability of the instructions to various user needs.

  ## Evaluation Process
  For each criterion:
  - Provide a score from 1-10
  - Include a brief justification explaining the rationale behind the score
  - Offer specific suggestions for improvement where applicable

  ## Additional Considerations
  - Are the instructions intuitive and easy to follow?
  - Do they account for users with different levels of technical expertise?
  - Are there clear examples or step-by-step guides?

  ## Output Format
  {OUTPUT_EXAMPLE}

  ## Final Recommendation
  Provide an overall assessment of the instruction's effectiveness and potential for guiding users in exploring the dataset.
  """

  return PROMPT
````

````python
import json

def print_json_content(json_response):
    json_response = json_response.replace("```json\n", "").replace("\n```", "")

    json_response_dict = json.loads(json_response)
    
    # Stampa ogni elemento del JSON in modo ordinato
    for key, value in json_response_dict.items():
        print(f"{key.replace('_', ' ').title()}:")
        
        # Se il valore è un dizionario, stampa i suoi sotto-elementi
        if isinstance(value, dict):
            for sub_key, sub_value in value.items():
                print(f"  {sub_key.replace('_', ' ').title()}: {sub_value}")
        else:
            print(f"  {value}")
        print()  # Riga vuota per separare gli elementi
````
