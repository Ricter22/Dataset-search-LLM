```python
import json

jsonl_path = f"benchmark/data_search_e_collection.jsonl"
data_search_e_collection = []
with open(jsonl_path, 'r', encoding='utf-8') as f:
    for line in f:
        data_search_e_collection.append(json.loads(line))

print(data_search_e_collection[0])
```

```python
import pandas as pd
import os

# Trova un file CSV nella cartella benchmark\datasets
datasets_dir = os.path.join('benchmark', 'datasets')
csv_files = [f for f in os.listdir(datasets_dir) if f.lower().endswith('.csv')]

if csv_files:
    csv_path = os.path.join(datasets_dir, csv_files[0])
    df = pd.read_csv(csv_path)
    print(f"Caricato: {csv_files[0]}")
    display(df.head())
else:
    print("Nessun file CSV trovato nella cartella benchmark\\datasets.")
```

```python
# Lista dei file nella cartella datasets
file_to_id = {}
for file_name in os.listdir(datasets_dir):
    # Cerca il file_name come data_filename in data_search_e_collection
    found_id = None
    for item in data_search_e_collection:
        for d in item.get('data', []):
            if d.get('data_filename') == file_name:
                found_id = item.get('id')
                break
        if found_id:
            break
    if found_id:
        file_to_id[file_name] = found_id

print(file_to_id)
```

```python
import json

# Percorso del file
topics_path = os.path.join('benchmark', 'data_search_2_e_train_topics.tsv')

# Estrai tutti gli id delle query in una lista
query_ids = []
with open(topics_path, 'r', encoding='utf-8') as f:
    for line in f:
        parts = line.strip().split('\t')
        if parts:
            query_ids.append(parts[0])

print(query_ids)
```

```python
csv_dataset_ids = []
with open(jsonl_path, 'r', encoding='utf-8') as f:
    for line in f:
        item = json.loads(line)
        if any(d.get('data_format', '').lower() == 'csv' for d in item.get('data', [])):
            csv_dataset_ids.append(item.get('id'))

print(csv_dataset_ids)
```

```python
from collections import defaultdict

# Analizza il file qrels e calcola la percentuale di dataset_id che fanno parte di csv_dataset_ids per ogni query_id


# Dizionario per raccogliere i dataset_id per ogni query_id
query_to_datasets = defaultdict(set)

with open(qrels_path, 'r', encoding='utf-8') as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) >= 2:
            query_id, dataset_id = parts[0], parts[1]
            query_to_datasets[query_id].add(dataset_id)

# Calcola la percentuale per ogni query_id
percentuali_csv = {}
for query_id, dataset_ids in query_to_datasets.items():
    total = len(dataset_ids)
    if total == 0:
        percentuali_csv[query_id] = 0.0
        continue
    csv_count = sum(1 for d in dataset_ids if d in csv_dataset_ids)
    percentuali_csv[query_id] = csv_count / total * 100

# Mostra i risultati
for query_id, percentuale in percentuali_csv.items():
    if percentuale > 50:
        print(f"{query_id}: {percentuale:.2f}% dei dataset_id sono CSV")
```

```python
# Estrai i query_id con percentuale di file csv > 90%
high_csv_queries = [qid for qid, perc in percentuali_csv.items() if perc > 90]

# Recupera gli id dei dataset associati a questi query_id
high_csv_datasets = set()
for qid in high_csv_queries:
    high_csv_datasets.update(query_to_datasets[qid])

print("Query ID con >90% CSV:", high_csv_queries)
print("Dataset ID associati:", list(high_csv_datasets))
```

```python
import os
import requests

# Crea la cartella di destinazione se non esiste
os.makedirs('benchmark/datasets_90_csv', exist_ok=True)

downloaded = 0

for dataset_id in high_csv_datasets:
    # Cerca il dataset corrispondente in data_search_e_collection
    item = next((x for x in data_search_e_collection if x.get('id') == dataset_id), None)
    if not item:
        print(f"Dataset {dataset_id} non trovato.")
        continue
    for d in item.get('data', []):
        if d.get('data_format', '').lower() == 'csv':
            data_url = d.get('data_url')
            if not data_url:
                continue
            filename = d.get('data_filename') or os.path.basename(data_url.split('?')[0])
            dest_path = os.path.join('benchmark', 'datasets_90_csv', filename)
            try:
                response = requests.get(data_url, timeout=30)
                response.raise_for_status()
                with open(dest_path, 'wb') as f:
                    f.write(response.content)
                print(f"Scaricato: {filename}")
                downloaded += 1
            except Exception as e:
                print(f"Errore scaricando {data_url}: {e}")
```

```python
# Conta quanti dataset per ciascuna query in high_csv_queries sono stati scaricati nella cartella benchmark/datasets_90_csv

downloaded_files = set(os.listdir(os.path.join('benchmark', 'datasets_90_csv')))

for qid in high_csv_queries:
    dataset_ids = query_to_datasets[qid]
    # Conta quanti file associati a questi dataset_id sono stati scaricati
    downloaded_count = 0
    for dataset_id in dataset_ids:
        # Cerca il dataset in data_search_e_collection
        item = next((x for x in data_search_e_collection if x.get('id') == dataset_id), None)
        if not item:
            continue
        for d in item.get('data', []):
            if d.get('data_format', '').lower() == 'csv':
                filename = d.get('data_filename') or os.path.basename(d.get('data_url', '').split('?')[0])
                if filename in downloaded_files:
                    downloaded_count += 1
                    break  # Conta solo una volta per dataset_id
    print(f"Query {qid}: {downloaded_count} dataset CSV scaricati su {len(dataset_ids)} totali")
```

```python
from pprint import pprint

# Crea una struttura dati che associa a ciascuna query_id i dataset_id relativi,
# indicando per ciascuno se è stato scaricato o meno

downloaded_files = set(os.listdir(os.path.join('benchmark', 'datasets_90_csv')))
query_datasets_status = {}

for qid in high_csv_queries:
    dataset_ids = query_to_datasets[qid]
    datasets_info = []
    for dataset_id in dataset_ids:
        # Cerca il dataset in data_search_e_collection
        item = next((x for x in data_search_e_collection if x.get('id') == dataset_id), None)
        if not item:
            continue
        for d in item.get('data', []):
            if d.get('data_format', '').lower() == 'csv':
                filename = d.get('data_filename') or os.path.basename(d.get('data_url', '').split('?')[0])
                status = 'scaricato' if filename in downloaded_files else 'non scaricato'
                datasets_info.append({'dataset_id': dataset_id, 'filename': filename, 'status': status})
                break  # Conta solo una volta per dataset_id
    query_datasets_status[qid] = datasets_info

# Esempio di visualizzazione per la prima query
# Aggiungi la query_id come campo in ciascun dizionario del risultato
for qid, datasets in query_datasets_status.items():
    for d in datasets:
        d['query_id'] = qid
pprint(query_datasets_status[high_csv_queries[0]])
```
